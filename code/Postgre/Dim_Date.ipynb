```python
import os
import time
import pandas as pd
import akshare as ak
import warnings
from datetime import timedelta, datetime
warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('expand_frame_repr', False)
pd.set_option('display.unicode.ambiguous_as_wide', True)
pd.set_option('display.unicode.east_asian_width', True)
```

```python
import datetime
import chinese_calendar

start_date = datetime.date(2024, 1, 1) # 2023-1-1
end_date = datetime.date(2024, 12, 31) # 2023-12-31
date_range = [start_date + datetime.timedelta(days=x) for x in range((end_date - start_date).days + 1)] # Generate Date range for 2023
trade_dates = [date for date in date_range if (chinese_calendar.is_workday(date) and date.weekday()>=0 and date.weekday()<=4)] # Filter Trading dates
print(trade_dates)
```

```python
df = pd.DataFrame(trade_dates, columns=['Date']) 
df['Date'] = pd.to_datetime(df['Date'])
df['Date_ID'] = df['Date'].apply(lambda x: x.strftime('%Y%m%d')).astype(int)
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['WeekDay'] = df['Date'].apply(lambda x: x.weekday() + 1)  
df['Quarter'] = df['Month'].apply(lambda x: (x-1)//3+1)
df = df[['Date_ID','Year','Month','Day','WeekDay','Quarter']]
df
```

```python
df.to_csv('Dim_Date_2024.csv',index=False)
```
