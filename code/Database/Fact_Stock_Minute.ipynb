```python
def GetMinuteID(x:str):
    dt = datetime.strptime(x, "%Y-%m-%d %H:%M:%S")
    result = int(dt.strftime("%Y%m%d%H%M"))
    return result
    
def check_data_exist(Start_Min:str, stockid:str):
    sql = "SELECT COUNT(fsm_id) from \"Trading\".Fact_Stock_Minute\
                where Minute_ID=" + str(GetMinuteID(Start_Min)) + " and Stock_ID=" + stockid+";"
    conn = psycopg2.connect(host="localhost", database="yuzhuoran", user="postgres", password="YZRyzr010306!")
    cur = conn.cursor()
    conn.commit()
    conn.rollback()
    conn.autocommit = True
    cur.execute(sql)
    col_names = [desc [0] for desc in cur.description]
    results = cur.fetchall()
    cur.close()
    return col_names,results
    
def update_minute_data(Current_Date:str):
    conn = psycopg2.connect(host="localhost", database="yuzhuoran", user="postgres", password="YZRyzr010306!")
    cur = conn.cursor()
    conn.commit()
    conn.rollback()
    conn.autocommit = True
    start_min = Current_Date + ' 09:30:00'
    end_min = Current_Date + ' 15:00:00'
    for CodeDes in code_list.values:
        ak_code, ak_name = CodeDes[1],CodeDes[2]
        print(Current_Date,ak_code,ak_name)
        col,result = check_data_exist(start_min, ak_code[2:])
        if result[0][0]>0:
            print("Exists")
            continue
        try:
            stock_zh_a_hist_min_em_df = ak.stock_zh_a_hist_min_em(symbol=ak_code[2:], start_date=start_min, end_date=end_min, period='1', adjust='')
        except Exception as e:
            print(e)
            continue
        if len(stock_zh_a_hist_min_em_df)==0:
            continue
        stock_zh_a_hist_min_em_df.rename(columns = {'':'Time','':'Open','':'Close','':'High','':'Low',
                                                    '':'Volume','':'Turnover','':'Latest'},inplace = True)
        stock_zh_a_hist_min_em_df['Minute_ID'] = stock_zh_a_hist_min_em_df['Time'].apply(lambda x:GetMinuteID(x))
        stock_zh_a_hist_min_em_df['FSM_ID'] = ak_code[2:]+stock_zh_a_hist_min_em_df['Minute_ID'].astype(str)
        stock_zh_a_hist_min_em_df['Stock_ID'] = ak_code[2:]
        stock_zh_a_hist_min_em_df['Stock_Name'] = ak_name
        stock_zh_a_hist_min_em_df = stock_zh_a_hist_min_em_df[['FSM_ID','Stock_ID','Minute_ID','Open',
                                                               'High','Low','Volume','Stock_Name','Turnover','Close']]
        sql = "INSERT INTO \"Trading\".Fact_Stock_Minute (FSM_ID, Stock_ID, Minute_ID, Open, High, Low, Volume, \
                                        Stock_Name, Turnover, Close) \
                                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        
        for index, row in stock_zh_a_hist_min_em_df.iterrows():
            values = (row['FSM_ID'], row['Stock_ID'],row['Minute_ID'],row['Open'], row['High'],row['Low'],row['Volume'],
                    row['Stock_Name'],row['Turnover'],row['Close']) 
            cur.execute(sql, values)
    cur.close()
```
