```python
def create_path(ak_code:str):
    global path
    path = os.path.join(".","Stock_Daily_2024")
    # save data
    if not os.path.exists(path) :
        # os.mkdir(path) 
        os.makedirs(path)  
    file_name = ak_code + ".csv"
    return os.path.join(path,file_name)
```

```python
def check(ak_code:str):
    path = os.path.join(".","Stock_Daily_2024",ak_code)
    path = path + ".csv"
    print(path)
    if os.path.isfile(path):
        return True
    return False
```

```python
code_list = pd.read_csv('CodeList.csv')
def check_sql_exist(Current_Date:str, stockid:str):
    sql = "SELECT COUNT(fsd_id) from \"Trading\".Fact_Stock_Daily\
                where Date_ID=" + Current_Date + " and Stock_ID=" + stockid+";"
    conn = psycopg2.connect(host="localhost", database="yuzhuoran", user="postgres", password="YZRyzr010306!")
    cur = conn.cursor()
    conn.commit()
    conn.rollback()
    conn.autocommit = True
    cur.execute(sql)
    col_names = [desc [0] for desc in cur.description]
    results = cur.fetchall()
    cur.close()
    return col_names,results
    
def update_daily_data(Current_Date:str):
    conn = psycopg2.connect(host="localhost", database="yuzhuoran", user="postgres", password="YZRyzr010306!")
    cur = conn.cursor()
    conn.commit()
    conn.rollback()
    conn.autocommit = True
    for CodeDes in code_list.values:
        ak_code, ak_name = CodeDes[1],CodeDes[2]
        print(ak_code,ak_name)
        col,result = check_sql_exist(Current_Date,ak_code[2:])
        if result[0][0] > 0:
            print(True)
            continue
        try:
            stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=ak_code[2:], start_date=Current_Date, end_date=Current_Date, period='daily', adjust='')
        except Exception as e:
            print(e)
            continue
        if len(stock_zh_a_hist_df)==0:
            continue
            #‘2023-12-01’
        stock_zh_a_hist_df['Date_ID'] = stock_zh_a_hist_df['日期'].apply(lambda x:x.strftime("%Y%m%d")).astype(int)
        stock_zh_a_hist_df['FSD_ID'] = stock_zh_a_hist_df['Date_ID']*1000000+int(ak_code[2:])
        stock_zh_a_hist_df['Stock_ID'] = int(ak_code[2:])
        stock_zh_a_hist_df.rename(columns = {'日期':'Date','开盘':'Open','收盘':'Close','最高':'High','最低':'Low',
                                                    '成交量':'Volume','成交额':'Turnover','振幅':'Amplitude',
                                             '涨跌幅':'Percent_Of_Incre_Decre','涨跌额':'Amount_Of_Incre_Decre','换手率':'Turnover_Rate'},inplace = True)
        stock_zh_a_hist_df = stock_zh_a_hist_df[['FSD_ID','Stock_ID','Date_ID','Open','Close','High','Low','Volume',
                                                 'Turnover','Amplitude','Percent_Of_Incre_Decre','Amount_Of_Incre_Decre','Turnover_Rate']]
        
        sql = "INSERT INTO \"Trading\".Fact_Stock_Daily (FSD_ID, Stock_ID, Date_ID, Open, Close, High, Low, Volume, \
                                        Turnover, Amplitude, Percent_Of_Incre_Decre, Amount_Of_Incre_Decre, Turnover_Rate) \
                                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        
        for index, row in stock_zh_a_hist_df.iterrows():
            values = (row['FSD_ID'], row['Stock_ID'],row['Date_ID'],row['Open'],row['Close'], row['High'],row['Low'],row['Volume'],
                    row['Turnover'],row['Amplitude'],row['Percent_Of_Incre_Decre'],row['Amount_Of_Incre_Decre'],row['Turnover_Rate']) 
            cur.execute(sql, values)
    cur.close()
    
```
